```python
import json
import numpy as np
import matplotlib.pyplot as plt

def perspective_camera_matrices(near, far, left, right, top, bottom, u, v, n, r):
    camera_mat = np.eye(4)
    camera_mat[:3, 0] = u
    camera_mat[:3, 1] = v
    camera_mat[:3, 2] = n
    camera_mat[:3, 3] = -r
    camera_mat[0, 3] = -np.dot(r, u)
    camera_mat[1, 3] = -np.dot(r, v)
    camera_mat[2, 3] = -np.dot(r, n)
    camera_mat[3, 3] = 1
    perspective_mat = np.zeros((4, 4))
    perspective_mat[0, 0] = 2 * near / (right - left)
    perspective_mat[1, 1] = 2 * near / (top - bottom)
    perspective_mat[0, 2] = (right + left) / (right - left)
    perspective_mat[1, 2] = (top + bottom) / (top - bottom)
    perspective_mat[2, 2] = -(far + near) / (far - near)
    perspective_mat[2, 3] = -2 * far * near / (far - near)
    perspective_mat[3, 2] = -1
    return camera_mat, perspective_mat
             
# Function to calculate line equation
def line_eq(v0, v1, x, y):
    return (v0[1] - v1[1]) * x + (v1[0] - v0[0]) * y + (v0[0] * v1[1] - v1[0] * v0[1])

def compute_triangle_color(normal):
    # Calculate the dot product of the normal with the light vector
    dotp = 0.707 * normal[0] + 0.5 * normal[1] + 0.5 * normal[2]
    # Make the dot product absolute if negative, or clamp to 1.0 if above
    dotp = abs(dotp) if dotp < 0 else (1.0 if dotp > 1.0 else dotp)
    # Calculate the triangle color, "tinting" the gray to a purplish hue
    tri_color = [0.95 * dotp, 0.65 * dotp, 0.88 * dotp]
    return tri_color

# Main rendering function with z-buffer
def render_with_z_buffer(triangles, canvas, z_buffer, raster_vertex):
    canvas_height, canvas_width, _ = canvas.shape
    for triangle, raster_vertex in zip(triangles, raster_vertices):
        v0, v1, v2 = raster_vertex[0], raster_vertex[1], raster_vertex[2]
        normal = np.array(triangle["v0"]["n"])
        color = compute_triangle_color(normal)

        # Calculate bounding box and clamp values
        xmin, xmax = np.floor(min(v0[0], v1[0], v2[0])), np.ceil(max(v0[0], v1[0], v2[0]))
        ymin, ymax = np.floor(min(v0[1], v1[1], v2[1])), np.ceil(max(v0[1], v1[1], v2[1]))
        xmin, xmax = max(xmin, 0), min(xmax, canvas_width - 1)
        ymin, ymax = max(ymin, 0), min(ymax, canvas_height - 1)

        # Precalculate denominators for efficiency
        f12_v0 = line_eq(v1, v2, v0[0], v0[1])
        f20_v1 = line_eq(v2, v0, v1[0], v1[1])
        f01_v2 = line_eq(v0, v1, v2[0], v2[1])

        for y in range(int(ymin), int(ymax) + 1):
            for x in range(int(xmin), int(xmax) + 1):
                alpha = line_eq(v1, v2, x, y) / f12_v0
                beta = line_eq(v2, v0, x, y) / f20_v1
                gamma = line_eq(v0, v1, x, y) / f01_v2

                if alpha >= 0 and beta >= 0 and gamma >= 0:
                    # Calculate z-depth using barycentric coordinates
                    z = alpha * v0[2] + beta * v1[2] + gamma * v2[2]
                    if z < z_buffer[y, x]:
                        canvas[y, x] = color  # Set color if z-depth is closer
                        z_buffer[y, x] = z  # Update z-buffer

# Constants
canvas_size = 256
canvas = np.zeros((canvas_size, canvas_size, 3))
near, far = 2, 4
left, right, top, bottom = -10, 10, 10, -10

# Load teapot data from JSON
with open('teapot.json', 'r') as f:
    teapot_data = json.load(f)

# Initialize image and z-buffer
z_buffer = np.full((canvas_size, canvas_size), np.inf, dtype=float)

# Define the camera space axes
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
n = np.array([0, 0, 1])
r = np.array([0, 0, 20])

# Create camera and perspective matrices
camera_mat, perspective_mat = perspective_camera_matrices(near, far, left, right, top, bottom, u, v, n, r)

vertices = []
raster_vertices = []

# Iterate over each triangle in the teapot data
for triangle in teapot_data['data']:
    camera_vertices = []
    # Transform vertices to camera space
    for vertex in [triangle['v0'], triangle['v1'], triangle['v2']]:
        v = np.array(vertex['v'][:3] + [1.0])  # Add 1.0 for 4D
        v_camera = camera_mat @ v
        camera_vertices.append(v_camera)

    ndc_vertices = []
    # Convert vertices to normalized device coordinates
    for v_camera in camera_vertices:
        v_ndc = v_camera[:3] / v_camera[3]
        v_ndc = np.append(v_ndc, 1)
        ndc_vertices.append(v_ndc)

    projected_vertices = []
    # Apply perspective projection to vertices
    for v_ndc in ndc_vertices:
        v_perspective = perspective_mat @ v_ndc
        projected_vertices.append(v_perspective)

    # Convert projected vertices to raster coordinates
    raster_vertex = []
    for v_perspective in projected_vertices:
        x = (v_perspective[0] + 1) * ((canvas_size - 1) / 2)
        y = (1 - v_perspective[1]) * ((canvas_size - 1) / 2)
        z = v_perspective[2]
        raster_vertex.append([x, y, z])
    raster_vertices.append(raster_vertex)

triangles = teapot_data["data"]
render_with_z_buffer(triangles, canvas, z_buffer, raster_vertices)

plt.imshow(canvas, cmap='gray')
plt.axis('off')  # Turn off axis numbers and ticks

# Save the plot as an image file
plt.savefig('teapot.png', bbox_inches='tight', pad_inches=0)

# Display the plot
plt.show()
```
